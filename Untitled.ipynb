```python
import pandas as pd
```

```python
df_domicilios = pd.read_csv('./dados/Domicilio01_SP1.csv',sep=';')
df_renda_domicilios = pd.read_csv('./dados/DomicilioRenda_SP1.csv',sep=';')
```

```python
df_domicilios = df_domicilios[['Cod_setor','V002']]
df_renda_domicilios['V003'] = df_renda_domicilios['V003'].str.replace('X','0').astype(int)
df_renda_domicilios = df_renda_domicilios[['Cod_setor','V003']]
df_renda_per_domi = pd.merge(left=df_domicilios,right=df_renda_domicilios,on='Cod_setor',how='inner')
```

```python
df_renda_per_domi['RENDA_PER_DOMPP'] = df_renda_per_domi['V003'] / df_renda_per_domi['V002']
```

```python
df_renda_per_domi.rename(columns={'V002':'DOMPP','V003':'RENDA_DOMPP'},inplace=True)
```
